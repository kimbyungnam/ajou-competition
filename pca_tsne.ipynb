```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
```

```python
training_data = np.loadtxt('./fmnist_train.csv', delimiter=",")
training_img = training_data[:,1:]
training_lab = training_data[:,0]

test_img = np.loadtxt('./fmnist_test.csv', delimiter=",")
```

```python
_ = scatter_matrix(pd.DataFrame(training_img), alpha=0.7,
                   c=training_lab, s=100, figsize=(12,12), diagonal="kde")
```

```python
pca_result = PCA(2).fit_transform(training_img)

#plt.scatter(pca_result[:,0], pca_result[:,1])
plt.scatter(pca_result[:,0], pca_result[:,1], c = training_lab)
```

```python
tsne_result = TSNE(n_components=2, init='pca').fit_transform(training_img)

plt.scatter(tsne_result[:,0], tsne_result[:,1], c = training_lab)
```
